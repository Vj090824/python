```python
Consider the below code to answer further questions:
import numpy as np
list_ = [ ‘1’ , ’2’ , ‘3’ , ‘4’ , ‘5’ ]
array_list = np.array(object = list_)
Q1. Is there any difference in the data type of variables list_ and array_list? 
If there is then write a code to print the data types of both the variables.





Ans:
    
    Yes, there is a difference in the data type of variables `list_` and `array_list`. 

The variable `list_` is a Python list, and its elements are of type string.
The variable `array_list` is a NumPy array, and its elements are of type string as well.

You can print the data types of both variables using the `type()` function in Python:


import numpy as np

list_ = ['1', '2', '3', '4', '5']
array_list = np.array(object=list_)

print("Data type of 'list_':", type(list_))
print("Data type of 'array_list':", type(array_list))


When you run this code, you will see the data types of both variables printed to the console.
They should both be of type `<class 'list'>` for `list_` and `<class 'numpy.ndarray'>`
for `array_list`.












Q2. Write a code to print the data type of each and every element of both the variables list_ and
arra_list.


Ans:
    
    
You can use a loop to iterate through the elements of both the `list_` 
and `array_list` and print their data types.
Here's a Python code snippet to do that:


import numpy as np

list_ = ['1', '2', '3', '4', '5']
array_list = np.array(list_)

# Print the data type of each element in list_
for element in list_:
    print(f"Element: {element}, Data Type: {type(element)}")

# Print the data type of each element in array_list
for element in array_list:
    print(f"Element: {element}, Data Type: {type(element)}")

This code will first iterate through the elements in the `list_` and then in the
`array_list`, printing the data type of each element. Note that the data types will 
differ between the two because the elements 
in the `list_` are strings, while the elements in `array_list`
will be of the data type `numpy.str_`.















Q3. Considering the following changes in the variable, array_list:
array_list = np.array(object = list_, dtype = int)
Will there be any difference in the data type of the elements present in both the variables, list_ and
arra_list? If so then print the data types of each and every element present in both the variables, list_
and arra_list.
Consider the below code to answer further questions:
import numpy as np
num_list = [ [ 1 , 2 , 3 ] , [ 4 , 5 , 6 ] ]
num_array = np.array(object = num_list)





Ans:
    
    
    In the given code snippet, you have a Python list called `num_list` containing nested lists with integers, 
    and you are converting it into a NumPy array using the following code:


import numpy as np
num_list = [[1, 2, 3], [4, 5, 6]]
num_array = np.array(object=num_list)


In this case, the elements in both `num_list` and `num_array` will have the 
same data type because you didn't explicitly specify a data type in the `np.array()`
function. NumPy will infer the data type from the elements present in the `num_list`,
and it will choose the most appropriate data type to represent those elements. In this case,
since all elements are integers, both `num_list` and `num_array` will have elements of data type `int`.

So, there won't be any difference in the data type of the elements present in both `num_list`
and `num_array`, and both will contain elements of data type `int`.

If you want to confirm the data types of the elements in both variables, you can use the 
`dtype` attribute of the NumPy array:


print("Data type of elements in num_list:", type(num_list[0][0]))
print("Data type of elements in num_array:", num_array.dtype)


This code will print the data type of elements in both variables,
and you'll see that they are both of type `int`.










Q4. Write a code to find the following characteristics of variable, num_array:
(i) shape
(ii) size



Ans:
    
    To find the shape and size of a NumPy array in Python, you can use the following code:


import numpy as np

# Create a sample NumPy array
num_array = np.array([[1, 2, 3], [4, 5, 6]])

# (i) Find the shape of the array
shape = num_array.shape

# (ii) Find the size of the array
size = num_array.size

print("Shape of num_array:", shape)
print("Size of num_array:", size)

In this code:
- We import the NumPy library as `np`.
- We create a sample NumPy array `num_array`.
- To find the shape of the array, we use the `shape` attribute of the NumPy array, 
which returns a tuple representing the dimensions of the array.
- To find the size of the array, we use the `size` attribute of the NumPy array, 
which returns the total number of elements in the array.
- We then print out the shape and size of the `num_array`.












Q5. Write a code to create numpy array of 3*3 matrix containing zeros only, using a numpy array
creation function.
[Hint: The size of the array will be 9 and the shape will be (3,3).]




Ans:
    
    You can create a numpy array of a 3x3 matrix containing zeros using the
    `numpy.zeros` function. Here's the code to do that:


import numpy as np

# Create a 3x3 matrix containing zeros
zeros_matrix = np.zeros((3, 3))

print(zeros_matrix)


This code will create a numpy array with a shape of (3, 3) filled with zeros.













Q6. Create an identity matrix of shape (5,5) using numpy functions?
[Hint: An identity matrix is a matrix containing 1 diagonally and other elements will be 0.]



Ans:
    
    
You can create an identity matrix of shape (5,5) using the `numpy` library in Python.
Here's how you can do it:


import numpy as np

identity_matrix = np.eye(5)
print(identity_matrix)


This code imports numpy as np and then uses the `np.eye()` function to create a 5x5 identity matrix.
The `np.eye()` function creates a square matrix with 1s on the diagonal and 0s elsewhere.
```
