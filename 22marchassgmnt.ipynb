```python
Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.


Ans:
    
 To calculate the Pearson correlation coefficient (often denoted as "r") between two variables, such as
the amount of time students spend studying for an exam and their final exam scores, you need to follow these steps:

1. Organize your data: Make a list of the paired data points, where each data point consists of a student's
study time and their corresponding exam score.

2. Calculate the mean (average) of each variable:
   - Calculate the mean of the study time.
   - Calculate the mean of the exam scores.

3. Calculate the deviation of each data point from the mean:
   - For each data point, subtract the mean of study time from the study time of that data point.
   - Similarly, subtract the mean of exam scores from the exam score of that data point.

4. Multiply the deviations of the two variables for each data point:
   - Multiply the deviation of study time for a data point by the deviation of exam scores for the same
data point. Do this for all data points and sum up these products.

5. Calculate the sum of the squared deviations of each variable:
   - Square the deviations of study time for all data points and sum them up.
   - Square the deviations of exam scores for all data points and sum them up.

6. Calculate the Pearson correlation coefficient (r):
-Divide the sum of the products of deviations (from step 4) by the square root of the product of the
sum of squared deviations for both variables (from step 5). The formula for r is:
     
        r =   ∑(x− xˉ)(y− yˉ)/{sqrt (∑(x− xˉ)^2  ∑(y− yˉ)^2)}
     

     Here's how to interpret the result:

- If r is close to 1 or -1, it indicates a strong linear relationship between the two variables.
A value of 1 means a perfect positive linear relationship, while -1
means a perfect negative linear relationship.
- If r is close to 0, it suggests a weak or no linear relationship between the two variables.

For example:
- If r is 0.8, it suggests a strong positive linear relationship, indicating that as students 
spend more time studying, their exam scores tend to increase.
- If r is -0.6, it suggests a moderate negative linear relationship, indicating that as students 
spend more time studying, their exam scores tend to decrease.

Keep in mind that correlation does not imply causation. A high correlation does not prove that one
variable causes the other; it simply means they are related in a linear manner. Other factors could
be at play, so further analysis may be needed to draw meaningful conclusions.   
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.


Ans:
    
    To calculate Spearman's rank correlation between two variables, you need to follow these steps:

1. Rank the data: Assign ranks to the values in each variable separately. If there are tied values,
assign an average rank to those tied values. You will have ranked datasets
for both sleep and job satisfaction.

2. Calculate the difference between the ranks for each pair of data points and square the differences.

3. Sum up the squared differences.

4. Use the formula for Spearman's rank correlation (ρ) to calculate the correlation coefficient:

  ρ = 1 - frac{6∑ d^2}/{n(n^2 - 1)} 

   Where:
   - d is the difference between the ranks of corresponding data points.
   - n is the number of data points.

Now, let's assume you have collected data for 10 individuals, and here are their 
sleep durations and job satisfaction levels:

| Individual | Sleep (hours) | Job Satisfaction |
|------------|---------------|------------------|
| 1          | 7             | 8                |
| 2          | 6             | 7                |
| 3          | 8             | 6                |
| 4          | 7             | 9                |
| 5          | 6             | 5                |
| 6          | 5             | 4                |
| 7          | 8             | 8                |
| 8          | 7             | 7                |
| 9          | 6             | 6                |
| 10         | 5             | 5                |

Let's rank the data for both sleep and job satisfaction:

Sleep Ranks:
- Individual 1: 4
- Individual 2: 6.5 (tied)
- Individual 3: 1
- Individual 4: 4
- Individual 5: 6.5 (tied)
- Individual 6: 10 (tied)
- Individual 7: 1
- Individual 8: 4
- Individual 9: 6.5 (tied)
- Individual 10: 10 (tied)

Job Satisfaction Ranks:
- Individual 1: 6.5 (tied)
- Individual 2: 4
- Individual 3: 1
- Individual 4: 9
- Individual 5: 7
- Individual 6: 4
- Individual 7: 6.5 (tied)
- Individual 8: 4
- Individual 9: 6.5 (tied)
- Individual 10: 7

Now, let's calculate Spearman's rank correlation (ρ):

    ρ = 1 - frac{6∑ d^2}/{n(n^2 - 1)}

Where d is the difference between the ranks of corresponding data points.

Let's calculate \(d^2\) for each pair of data points, sum them up,
and calculate the correlation coefficient:

    ∑d^2 = (4-6.5)^2 + (6.5-4)^2 + (1-1)^2 + (4-9)^2 + (6.5-7)^2 + (10-4)^2 +
    (1-6.5)^2 + (4-4)^2 + (6.5-6.5)^2 + (10-7)^2 

    ∑d^2 = 82.5 

Now, plug this into the formula:

ρ = 1 - frac{6 × 82.5}{10(10^2 - 1)} 

ρ = 1 - frac{495}{990} 

 ρ = 1 - 0.5 

ρ = 0.5 

Interpretation:
The Spearman's rank correlation coefficient (\(ρ\)) between the amount of sleep individuals
get each night and their overall job satisfaction level is 0.5. This indicates a moderate positive 
monotonic relationship between these two variables. In other words, as the amount of sleep increases,
job satisfaction tends to increase as well, and as the amount of sleep decreases, job satisfaction
tends to decrease. However,
keep in mind that correlation does not imply causation, so other factors may also
be influencing job satisfaction. 
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
Q3. Suppose you are conducting a study to examine the relationship between the number of hours of
exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables
for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation
between these two variables and compare the results.




Ans:


To calculate the Pearson correlation coefficient and Spearman's rank correlation coefficient between
the number of hours of exercise per week and body mass index (BMI) in a sample of 50 adults,
you can follow these steps:

**Step 1: Organize Your Data**
Ensure you have a dataset that includes the number of hours of exercise per week (X) and the
corresponding BMI values (Y) for all 50 participants.

**Step 2: Calculate Pearson Correlation Coefficient (r)**

The Pearson correlation coefficient is used to measure the linear relationship between two
continuous variables. You can calculate it using the following formula:

    r = frac{∑((X_i - xˉ)(Y_i - yˉ ))}/{sqrt{∑(X_i - xˉ)^2 ∑(Y_i - yˉ )^2}} 

Where:
- X_i and Y_i are the individual data points for exercise hours and BMI.
- xˉ and  yˉ  are the means of exercise hours and BMI, respectively.

**Step 3: Calculate Spearman's Rank Correlation Coefficient (ρ)**

Spearman's rank correlation coefficient assesses the strength and direction of the monotonic
relationship between two variables. You will need to rank the data points for both variables
and then calculate the correlation. Here's how you do it:

1. Rank the data for both variables separately.
2. Calculate the differences in ranks (d) between the paired data points (d = rank(X) - rank(Y)).
3. Square each of these differences (d^2).
4. Calculate the Spearman's rank correlation coefficient using the formula:

   ρ = 1 - frac{6∑ d^2}/{n(n^2 - 1)}

Where:
- d is the difference in ranks.
- n is the number of data points (50 in this case).

**Step 4: Compare the Results**

Now that you've calculated both the Pearson correlation coefficient (r) and Spearman's
rank correlation coefficient (ρ), you can compare the results:

- If both r and ρ are close to 1 or -1, it suggests a strong monotonic and linear relationship, respectively.
- If r is close to 0 and ρ is not, it suggests a nonlinear relationship.
- If both r and ρ are close to 0, it suggests no significant relationship between the two variables.

Remember that the interpretation of the correlation coefficient should take into account the context
of the study and the characteristics of the data.

Perform the calculations based on your dataset to obtain the actual correlation 
coefficients and make the comparison.

we can compare the results to assess the relationship between exercise hours and BMI.
    
    
    
    
    







    




Q4. A researcher is interested in examining the relationship between the number of hours individuals
spend watching television per day and their level of physical activity. The researcher collected data on
both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between
these two variables.



Ans:

To calculate the Pearson correlation coefficient (also known as Pearson's r) between the number
    of hours individuals spend watching television per day and their level of physical activity, 
you'll need the following information:

The raw data for both variables (hours of TV watching and level of physical activity)
for the 50 participants.
Once you have this data, you can use the following formula to calculate Pearson's correlation coefficient:

1. Compute the mean (average) of both variables: the number of hours spent watching television per day (X)
    and the level of physical activity (Y).

2. Calculate the deviation of each data point from its respective mean for both X and Y.

3. Multiply these deviations for each data point (X and Y).

4. Calculate the sum of the products from step 3.

5. Calculate the square root of the sum of the squared deviations of X and Y from their respective means.

6. Divide the result from step 4 by the result from step 5 to get the Pearson correlation coefficient (r).

The formula for Pearson's r is as follows:

     r = frac{∑((X_i - xˉ)(Y_i - yˉ ))}/{sqrt{∑(X_i - xˉ)^2 ∑(Y_i - yˉ )^2}} 

Where:
- X_irepresents each data point for the number of hours spent watching television.
- Y_i represents each data point for the level of physical activity.
- xˉ is the mean of the number of hours spent watching television.
- yˉ is the mean of the level of physical activity.

Here are the steps to calculate the Pearson correlation coefficient:

Calculate the mean (average) of the number of hours spent watching TV (let's call this variable X)
and the mean of the level of physical activity (let's call this variable Y) from your data.

For each participant, calculate the deviation from the mean for both X and Y. Subtract the mean of
X from each X value and subtract the mean of Y from each Y value.

Square each of these deviations calculated in step 2.

Multiply the deviations for X and Y for each participant and sum these products.

Calculate the square root of the sum of the squared deviations for X and the sum 
of the squared deviations for Y.

Divide the result from step 4 by the result from step 5 to get the Pearson correlation coefficient (r).

Here's a step-by-step example with hypothetical data:

Let's say we have the following data for 50 participants:

Hours of TV watching (X): [2, 3, 1, 4, 2, ...]
Physical activity level (Y): [5, 4, 6, 3, 5, ...]
Calculate the means:

Xˉ (mean of X) = Sum of all X values / Number of participants

Yˉ  (mean of Y) = Sum of all Y values / Number of participants

Calculate deviations for X and Y for each participant:
Deviation for X = X_i - Xˉ


 
Deviation for Y = Y_i - Yˉ

 

Square each deviation:
(Deviation for X)^2
(Deviation for Y)^2

4.Multiply deviations for X and Y and sum these products for all participants:

∑(DeviationforX∗DeviationforY)

5.Calculate the square root of the sum of squared deviations for X and Y:
 sqrt{∑(Deviation for X)^2 ∗ (Deviation for Y)^2


 Since you have data for 50 participants, you'll need to calculate all these values
using the data. Once you have these values, 
plug them into the formula to compute the Pearson correlation coefficient (r)

Divide the result from step 4 by the result from step 5 to get the Pearson correlation coefficient (r).
It important to note that this calculation can be quite labor-intensive when done manually 
for a large dataset.
This will give you the strength and direction of the linear relationship between the number of hours
individuals spend watching television per day and their level of physical activity. 
The value of r will range from -1 (perfect negative correlation) to 1 (perfect positive correlation),
with 0 indicating no linear correlation.
    
    
    
    
    






Q5. A survey was conducted to examine the relationship between age and preference for a particular
brand of soft drink. The survey results are shown below:

Age(Years)                 Soft drink Preference
25                            Coke
42                            Pepsi
37                            Mountain dew
19                            Coke
31                            Pepsi
28                            Coke



Ans:
    


A partial dataset from a survey examining the relationship between 
age and preference for a particular brand of soft drink. To better understand and analyze the relationship,
you would typically need a larger dataset with more participants. However, with the limited data 
you've provided, you can still make some basic observations:

1. **Age and Soft Drink Preference**: You can see that each row in the dataset represents an 
individual's age and their soft drink preference.

2. **Age Range**: The ages in the dataset range from 19 to 42 years.

3. **Soft Drink Preferences**: The soft drink preferences in the dataset are 
      "Coke," "Pepsi," and "Mountain Dew."

4. **Individual Preferences**:
   - Person 1 (Age 25) prefers "Coke."
   - Person 2 (Age 42) prefers "Pepsi."
   - Person 3 (Age 37) prefers "Mountain Dew."
   - Person 4 (Age 19) prefers "Coke."
   - Person 5 (Age 31) prefers "Pepsi."
   - Person 6 (Age 28) prefers "Coke."

To analyze the relationship between age and soft drink preference, you can start by calculating some 
basic descriptive statistics. You may want to calculate measures of central tendency (mean, median, mode)
for the ages and see if there are any patterns in the soft drink preferences.
      Additionally, you can create a frequency table to count how many people prefer each soft drink.

Here's a simple analysis based on the provided data:

Descriptive Statistics for Age:

Mean age: (25 + 42 + 37 + 19 + 31 + 28) / 6 = 30.33 years (approximately)
Median age: Since there are six data points, the median is the average of the 3rd and 4th values,
which are 28 and 31. So, the median age is (28 + 31) / 2 = 29.5 years (approximately)
Soft Drink Preferences Frequency:

Coke: 3 people
Pepsi: 2 people
Mountain Dew: 1 person
Based on this small sample, it appears that Coke is the most preferred soft drink among the survey 
participants, followed by Pepsi and Mountain Dew.

To explore the relationship further, you might consider visualizing the data with graphs or
conducting statistical tests to determine if there's a significant correlation between age 
and soft drink preference. However, with only six data points, it's important to acknowledge
that the results may not be highly statistically significant, and this data might not represent
the broader population's preferences accurately.
















Q6. A company is interested in examining the relationship between the number of sales calls made per day
and the number of sales made per week. The company collected data on both variables from a sample of
30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.




Ans:
    
 To calculate the Pearson correlation coefficient (often denoted as "r") between the number of sales calls 
    made per day and the number of sales made per week, you'll need the following information:

1. The data for the number of sales calls made per day.
2. The data for the number of sales made per week.

Once you have this data, you can use the following formula to calculate the Pearson correlation coefficient:

 r = frac{∑((X_i - xˉ)(Y_i - yˉ ))}/{sqrt{∑(X_i - xˉ)^2 ∑(Y_i - yˉ )^2}} 

Where:
- X_i represents each value in the sales calls made per day.
- Y_i represents each value in the sales made per week.
-  xˉis the mean (average) of the sales calls made per day.
- yˉ is the mean (average) of the sales made per week.

You'll need to calculate the following:

1. Calculate the mean () of the sales calls made per day.
2. Calculate the mean (yˉ) of the sales made per week.
3. Calculate the numerator, which is the sum of the product of the differences between each data point
and the respective mean for both variables.
4. Calculate the denominator, which is the square root of the product of the sums of the squared
differences between each data point and the respective mean for both variables.

Here's a step-by-step example calculation:

Assuming you have the following data for 30 sales representatives:

Sales Calls Made per Day (X):
- X1 = 10
- X2 = 15
- ...
- X30 = 12

Sales Made per Week (Y):
- Y1 = 5
- Y2 = 7
- ...
- Y30 = 6

1. Calculate the means:
xˉ = (10 + 15 + ... + 12) / 30
yˉ = (5 + 7 + ... + 6) / 30

2. Calculate the numerator:
Numerator = Σ[(Xi - xˉ )(Yi - yˉ)] for i = 1 to 30

3. Calculate the denominators:
Denominator_X = √[Σ(Xi - xˉ)^2] for i = 1 to 30
Denominator_Y = √[Σ(Yi - yˉ )^2] for i = 1 to 30

4. Finally, calculate the Pearson correlation coefficient (r):
r = frac{Numerator}{DenominatorX * DenominatorY}

This will give you the Pearson correlation coefficient, which quantifies the strength and direction
of the linear relationship between the two variables. The value of r will be between -1 
(perfect negative correlation) and +1 (perfect positive correlation),
with 0 indicating no linear correlation.
      
```
