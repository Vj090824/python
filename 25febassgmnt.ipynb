````python
Consider following code to answer further questions:
import pandas as pd
course_name = [‘Data Science’, ‘Machine Learning’, ‘Big Data’, ‘Data Engineer’]
duration = [2,3,6,4]
df = pd.DataFrame(data = {‘course_name’ : course_name, ‘duration’ : duration})
Q1. Write a code to print the data present in the second row of the dataframe, df.



Ans:
    
    You can print the data present in the second row of the DataFrame `df` using the `iloc` 
    indexer, which allows you to access rows by their integer index. In Python, indexing starts at 0, 
so the second row has an index of 1. Here's the code to print the second row of the DataFrame:


second_row = df.iloc[1]
print(second_row)


This code will output:


course_name    Machine Learning
duration                      3
Name: 1, dtype: object


It displays the data for the second row of the DataFrame, including the
`course_name` and `duration` columns











Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?


Ans:
    
    In pandas, both the `loc` and `iloc` functions are used to access and manipulate data in 
    a DataFrame, but they have different ways of selecting data based on the index and labels.

1. `loc` (Label-based Selection):
   - `loc` is primarily used for label-based indexing. It allows you to select rows and columns
by their labels (i.e., by their index names or column names).
   - When using `loc`, you specify the row and column labels you want to select, and it returns
    the data based on those labels.
   - It includes the end label in the range when slicing. For example, `df.loc['A':'C']` will 
include rows 'A', 'B', and 'C'.

Example:

import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['X', 'Y', 'Z'])

# Select a single element by label
value = df.loc['X', 'A']

# Select a single column by label
column = df.loc[:, 'B']

# Select rows by label range
rows = df.loc['X':'Y']

# Select rows and specific columns by label
subset = df.loc['X':'Y', ['A', 'B']]


2. `iloc` (Integer-based Selection):
   - `iloc` is used for integer-based indexing. It allows you to select rows and columns
by their integer positions, regardless of the labels.
   - When using `iloc`, you specify the row and column positions you want to select,
    and it returns the data based on those positions.
   - It does not include the end position in the range when slicing. For example, 
`df.iloc[0:2]` will include rows at positions 0 and 1.

Example:

import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# Select a single element by integer position
value = df.iloc[0, 0]

# Select a single column by integer position
column = df.iloc[:, 1]

# Select rows by integer position range
rows = df.iloc[0:2]

# Select rows and specific columns by integer positions
subset = df.iloc[0:2, [0, 1]]


In summary, the key difference between `loc` and `iloc` is the way they select data:
- `loc` uses label-based selection, where you specify row and column labels.
- `iloc` uses integer-based selection, where you specify row and column positions.

Your choice between `loc` and `iloc` should depend on whether you want to access
data by labels or by integer positions.

















Q3. Reindex the given dataframe using a variable, reindex = [3,0,1,2] and store it in the variable, new_df
then find the output for both new_df.loc[2] and new_df.iloc[2].


Did you observe any difference in both the outputs? If so then explain it.
Consider the below code to answer further questions:
import pandas as pd
import numpy as np
columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1,2,3,4,5,6]
#Creating a dataframe:
df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)



Ans:
    
    To reindex the given DataFrame `df1` using the provided `reindex` list and store it
    in the variable `new_df`, you can use the `reindex` method in pandas as follows
    
import pandas as pd
import numpy as np

# Original DataFrame
columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1, 2, 3, 4, 5, 6]
df1 = pd.DataFrame(np.random.rand(6, 6), columns=columns, index=indices)

# Reindexing using the provided list
reindex = [3, 0, 1, 2]
new_df = df1.reindex(reindex)

# Output for new_df.loc[2] and new_df.iloc[2]
output_loc = new_df.loc[2]
output_iloc = new_df.iloc[2]

print("Output for new_df.loc[2]:\n", output_loc)
print("\nOutput for new_df.iloc[2]:\n", output_iloc)


Now, let's analyze the difference between `new_df.loc[2]` and `new_df.iloc[2]`:

1. `new_df.loc[2]`: This code retrieves the row with label '2' from the `new_df` DataFrame. 
Since `reindex` has reordered the rows according to `[3, 0, 1, 2]`, 
it will return the row that originally had the label '3' in the original DataFrame `df1`.

2. `new_df.iloc[2]`: This code retrieves the row at the integer index position '2' in the 
`new_df` DataFrame. Since `reindex` has reordered the rows according to `[3, 0, 1, 2]`,
it will return the row that originally had the integer index position '1' in the original DataFrame `df1`.

So, there is a difference between the outputs of `new_df.loc[2]` and `new_df.iloc[2]`.
`new_df.loc[2]` refers to the row with label '3' from the original DataFrame, while
`new_df.iloc[2]` refers to the row with integer index position '1' from the original DataFrame.


















Q4. Write a code to find the following statistical measurements for the above dataframe df1:
(i) mean of each and every column present in the dataframe.
(ii) standard deviation of column, ‘column_2’


Ans:
    
    
import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# (i) Mean of each column
column_means = df.mean()
print("Mean of each column:")
print(column_means)

# (ii) Standard deviation of the 'duration' column
duration_std = df['duration'].std()
print("\nStandard deviation of 'duration' column:", duration_std)



This code will print the mean of each column in the DataFrame df and the standard
deviation of the 'duration' column.















Q5. Replace the data present in the second row of column, ‘column_2’ by a string variable then find the
mean of column, column_2.
If you are getting errors in executing it then explain why.
[Hint: To replace the data use df1.loc[] and equate this to string data of your choice.]



Ans:


To replace the data in the second row of the 'column_2' column in a DataFrame,
you can use the `loc` method. Here's how you can do it in Python using the pandas library:


import pandas as pd

# Create a DataFrame (you should have your actual DataFrame)
data = {'column_1': [1, 2, 3, 4],
        'column_2': [5, 6, 7, 8]}
df1 = pd.DataFrame(data)

# Replace the data in the second row of 'column_2' with a string variable
df1.loc[1, 'column_2'] = 'your_string_variable_here'

# Calculate the mean of 'column_2'
mean_column_2 = df1['column_2'].mean()

print(df1)
print("Mean of column_2:", mean_column_2)


Replace `'your_string_variable_here'` with the string you want to use.
This code will replace the value in the second row of 'column_2' with the specified
string and then calculate the mean of 'column_2'.

If you encounter any errors while executing this code, it could be due to various reasons:

1. If your DataFrame `df1` is not defined correctly or doesn't contain the 'column_2' column,
you will get a KeyError. Make sure your DataFrame structure matches the example I provided.

2. If the second row in 'column_2' contains a data type that cannot be replaced with a string, 
you might encounter a data type error. Ensure that the data types are compatible.

3. If you haven't imported the pandas library using `import pandas as pd`, you will get an ImportError.

4. If your Python environment lacks the pandas library, you need to install
it using `pip install pandas`.

Make sure to address these potential issues based on your specific DataFrame and data to avoid errors.

    
    
    
    
    
    
    
    
    
    
    
    
    
    


Q6. What do you understand about the windows function in pandas and list the types of windows
functions?

Ans:
    
    
    
    In pandas, the term "window functions" generally refers to a group of functions that 
    allow you to perform various operations on a set of data points within a defined window
    or window size. These functions are primarily used in conjunction with the `DataFrame`
    or `Series` objects and are often employed in data analysis and manipulation tasks, 
    such as time series analysis and rolling calculations.

Here are some common types of window functions in pandas:

1. **Rolling Functions**: Rolling functions allow you to compute statistics, aggregates, 
or other operations over a rolling window of data points. You can specify the window size, 
and the function calculates the result for each window as it moves through the data.
Common rolling functions include:
   - `rolling().mean()`: Calculate the rolling mean.
   - `rolling().sum()`: Calculate the rolling sum.
   - `rolling().min()`: Calculate the rolling minimum.
   - `rolling().max()`: Calculate the rolling maximum.

2. **Expanding Functions**: Expanding functions compute the cumulative results over the
entire data up to the current point. They do not use a fixed-size window but rather include
all data up to the current row. Common expanding functions include:
   - `expanding().mean()`: Calculate the expanding mean.
   - `expanding().sum()`: Calculate the expanding sum.

3. **Exponentially Weighted Functions**: These functions calculate exponentially weighted 
statistics for a given data series. They assign exponentially decreasing weights to older
data points, giving more importance to recent data. Common exponentially weighted functions include:
   - `ewm().mean()`: Calculate the exponentially weighted moving average.
   - `ewm().var()`: Calculate the exponentially weighted moving variance.

4. **Groupby and Window Functions**: These functions combine the grouping and window operations. 
They allow you to apply window functions within each group of data.
Common groupby and window functions include:
   - `groupby().rolling().mean()`: Calculate rolling means within each group.
   - `groupby().expanding().sum()`: Calculate expanding sums within each group.

5. **Shifting Functions**: Shifting functions allow you to shift data points within
a Series or DataFrame. You can use these functions to create lead or lag values in your data.
   - `shift()`: Shift the data by a specified number of periods.

6. **Ranking Functions**: Ranking functions assign a rank to each element in a group or
window based on a specified criterion. Common ranking functions include:
   - `rank()`: Assign ranks to elements in a Series.

These window functions are valuable for performing time-series analysis, 
calculating moving averages, identifying trends, and conducting various types of data 
transformations and aggregations within specific windows of your data.
They provide a powerful toolset for data analysts and data scientists working with 
pandas and time-series data.
















Q7. Write a code to print only the current month and year at the time of answering this question.
[Hint: Use pandas.datetime function]




Ans:
    

    
    To print only the current month and year using the `pandas` library in Python,
    you can follow these steps:

1. Import the `pandas` library.
2. Get the current date and time.
3. Extract the month and year from the current date.
4. Print the extracted month and year.

Here's the Python code to accomplish this:


import pandas as pd
from datetime import datetime

# Get the current date and time
current_date = datetime.now()

# Extract the month and year
current_month = current_date.month
current_year = current_date.year

# Print the current month and year
print(f"Current Month: {current_month}")
print(f"Current Year: {current_year}")


This code first imports the necessary libraries, gets the current date and time using `datetime.now()`, 
and then extracts the month and year from the current date.
Finally, it prints the current month and year.
    
    
    
Current Month: 9
Current Year: 2023    
    
    
    
    
    
    
    
    
    
    
    


Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. 
The program should prompt the user to enter the dates and display the result.




Ans:


To create a Python program that calculates the difference between two dates in days, hours, and minutes using Pandas timedelta, you can follow these steps:

1. Import the necessary libraries, including Pandas.
2. Prompt the user to input two dates in the format YYYY-MM-DD.
3. Parse the user input into Pandas datetime objects.
4. Calculate the timedelta between the two dates.
5. Extract the days, hours, and minutes from the timedelta.
6. Display the result to the user.

Here's the Python program:

```python
import pandas as pd

# Function to get and validate a date from the user
def get_valid_date(prompt):
    while True:
        try:
            date_str = input(prompt)
            date = pd.to_datetime(date_str)
            return date
        except ValueError:
            print("Invalid date format. Please use YYYY-MM-DD.")

# Get the first date from the user
start_date = get_valid_date("Enter the first date (YYYY-MM-DD): ")

# Get the second date from the user
end_date = get_valid_date("Enter the second date (YYYY-MM-DD): ")

# Calculate the timedelta between the two dates
time_difference = end_date - start_date

# Extract days, hours, and minutes
days = time_difference.days
hours, remainder = divmod(time_difference.seconds, 3600)
minutes = remainder // 60

# Display the result
print(f"Time difference between the two dates:")
print(f"Days: {days} days")
print(f"Hours: {hours} hours")
print(f"Minutes: {minutes} minutes")


This program prompts the user for two dates,
calculates the difference using Pandas timedelta,
and then displays the result in days, hours, and minutes.

    
    
    Enter the first date (YYYY-MM-DD):  20220822
Enter the second date (YYYY-MM-DD):  20230924
Time difference between the two dates:
Days: 398 days
Hours: 0 hours
Minutes: 0 minutes





    
    
    
    
    
    



Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.



Ans:
    
    To achieve this task, we can use the `pandas` library in Python. Here's
a Python program that reads a CSV file, converts a specified column to a categorical 
data type, and displays the sorted data:


import pandas as pd

# Prompt the user for the file path
file_path = input("Enter the CSV file path: ")

# Prompt the user for the column name
column_name = input("Enter the column name to convert to categorical: ")

# Prompt the user for the category order (comma-separated)
category_order = input("Enter the category order (comma-separated): ").split(',')

# Read the CSV file into a DataFrame
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"File not found: {file_path}")
    exit(1)
except Exception as e:
    print(f"An error occurred while reading the CSV file: {str(e)}")
    exit(1)

# Convert the specified column to a categorical data type with the given order
try:
    df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)
except KeyError:
    print(f"Column not found: {column_name}")
    exit(1)
except Exception as e:
    print(f"An error occurred while converting the column to categorical: {str(e)}")
    exit(1)

# Sort the DataFrame by the categorical column
sorted_df = df.sort_values(by=[column_name])

# Display the sorted data
print("\nSorted Data:")
print(sorted_df)


Make sure you have the `pandas` library installed before running this program.
You can install it using `pip` if you haven't already:


pip install pandas


This program will prompt the user to enter the CSV file path, 
the column name to convert to categorical,  and the desired category order.
It will then display the sorted data based on the categorical column. 












Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.



Ans:
    
    To create a Python program that reads a CSV file containing sales data and visualizes
it using a stacked bar chart, you can use the `pandas` library for data manipulation and 
`matplotlib` for data visualization. Make sure to install these libraries if you haven't already:


pip install pandas matplotlib


Here's a Python program to accomplish this task:


import pandas as pd
import matplotlib.pyplot as plt

# Function to read and visualize CSV data
def visualize_sales_data(file_path):
    try:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)

        # Ensure the DataFrame contains at least 'Category', 'Date', and 'Sales' columns
    if 'Category' not in df.columns or 'Date' not in df.columns or 'Sales' not in df.columns:
        print("The CSV file must contain 'Category', 'Date', and 'Sales' columns.")
            return

        # Convert the 'Date' column to a datetime type
        df['Date'] = pd.to_datetime(df['Date'])

        # Group the data by 'Category' and 'Date' and calculate the total sales
        df_grouped = df.groupby(['Category', 'Date'])['Sales'].sum().unstack()

        # Create a stacked bar chart
        ax = df_grouped.plot(kind='bar', stacked=True, figsize=(12, 6))
        plt.xlabel('Date')
        plt.ylabel('Total Sales')
        plt.title('Stacked Bar Chart of Sales by Product Category Over Time')
        plt.legend(title='Category', loc='upper right')
        plt.show()

    except Exception as e:
        print(f"An error occurred: {str(e)}")

if __name__ == "__main__":
    file_path = input("Enter the path to the CSV file: ")
    visualize_sales_data(file_path)


Save this code to a Python file (e.g., `sales_visualizer.py`) and run it.
It will prompt you to enter the file path of your CSV data containing sales information. 
After entering the path, it will generate and display the stacked bar chart showing
sales by product category over time.

Make sure your CSV file follows the format with columns 'Category', 'Date', and 'Sales' 
for this program to work correctly. 


















Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write
a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and
displays the results in a table.
The program should do the followingM
I Prompt the user to enter the file path of the CSV file containing the student dataR
I Read the CSV file into a Pandas DataFrameR
I Calculate the mean, median, and mode of the test scores using Pandas toolsR
I Display the mean, median, and mode in a table.
Assume the CSV file contains the following columnsM
I Student ID: The ID of the studentR
I Test Score: The score of the student's test.
Example usage of the program:
Enter the file path of the CSV file containing the student data: student_data.csv
+-----------+--------+
| Statistic | Value |
+-----------+--------+
| Mean | 79.6 |
| Median | 82 |
| Mode | 85, 90 |
+-----------+--------+
Assume that the CSV file student_data.csv contains the following data:
Student ID,Test Score
1,85
2,90
3,80
4,75
5,85
6,82
7,78
8,85
9,90
10,85
The program should calculate the mean, median, and mode of the test scores and display the results
in a table.




Ans:
    
    
    
    
    You can accomplish this task using Python and the Pandas library.
    Make sure you have Pandas installed (`pip install pandas`) before running the program.
    Here's a Python program that reads the CSV file, calculates the mean, median, 
    and mode of the test scores, and displays the results in a table:

import pandas as pd

# Prompt the user to enter the file path of the CSV file
file_path = input("Enter the file path of the CSV file containing the student data: ")

# Read the CSV file into a Pandas DataFrame
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print("File not found. Please provide a valid file path.")
    exit(1)

# Calculate the mean, median, and mode of the test scores
mean_score = df['Test Score'].mean()
median_score = df['Test Score'].median()
mode_scores = df['Test Score'].mode()

# Display the results in a table
print("+-----------+--------+")
print("| Statistic | Value  |")
print("+-----------+--------+")
print(f"| Mean      | {mean_score:.1f}   |")
print(f"| Median    | {median_score}     |")
print(f"| Mode      | {', '.join(map(str, mode_scores))}   |")
print("+-----------+--------+")


Save this code in a Python file (e.g., `calculate_statistics.py`) and run it. 
Make sure your CSV file has the same format as specified in your example usage.
    
    
    
    
    
    
    
    
    
    
    
    
````
