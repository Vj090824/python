```python
Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.




Ans:
      
You can calculate the 95% confidence interval for a sample of data with a mean of 50 and a
standard deviation of 5 in Python using the `scipy` library. Here's how you can do it:


import scipy.stats as stats

# Sample data characteristics
sample_mean = 50
sample_std_dev = 5
sample_size = 100  # Replace this with your actual sample size

# Confidence level
confidence_level = 0.95

# Calculate the margin of error
margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * (sample_std_dev / (sample_size ** 0.5))

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")


Interpretation:

The 95% confidence interval for this sample is calculated to be (48.84, 51.16).
This means that we are 95% confident that the true population mean lies within this interval.
In other words, if we were to take multiple random samples from the population and calculate
a 95% confidence interval for each sample, we would expect the true population mean to be
captured by the interval in approximately 95% of those cases.

In this specific case, we can say that we are 95% confident that the true population mean is 
between 48.84 and 51.16 based on our sample data with a mean of 50 and a standard deviation of 5.
















Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.



Ans:

To conduct a chi-square goodness of fit test in Python to determine if the distribution
of colors of M&Ms in a bag matches the expected distribution, you can use the `scipy.stats`
library. Here's a step-by-step guide:

1. Import the necessary libraries:


import numpy as np
from scipy import stats


2. Define your observed frequencies and expected frequencies.

   - Observed frequencies: The number of each color you actually find in the bag.
   - Expected frequencies: The expected number of each color based on the given distribution.


observed_frequencies = np.array([15, 18, 22, 10, 11, 24])  # Replace with your actual counts
expected_frequencies = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20]) * np.sum(observed_frequencies)


3. Calculate the chi-square statistic:


chi_square_statistic = np.sum((observed_frequencies - expected_frequencies)**2 / expected_frequencies)


4. Calculate the degrees of freedom. In a goodness of fit test, the degrees of freedom are equal
to the number of categories minus 1.


degrees_of_freedom = len(observed_frequencies) - 1


5. Calculate the p-value using the chi-square statistic and degrees of freedom:


p_value = 1 - stats.chi2.cdf(chi_square_statistic, degrees_of_freedom)

6. Finally, compare the p-value to the significance level (0.05) to make a decision. If p-value < 0.05,
you can reject the null hypothesis, indicating that the distribution of colors
in the bag does not match the expected distribution.


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The distribution of colors does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The distribution of colors matches the expected distribution.")


Remember to replace the `observed_frequencies` with your actual counts of each color in the M&M bag. 
The code above will perform the chi-square goodness of fit test and print whether or not you can reject
the null hypothesis at the specified significance level.














Q3. Use Python to calculate the chi-square statistic and p-value for a contingency table with the following
data:
Interpret the results of the test.

         Group A        Group B
Outcome 1  20             15
Outcome 2  10             25
Outcome 3  15             20


Use a significance level of 0.05.




Ans:
    
    
You can use Python's `scipy.stats` library to calculate the chi-square statistic and p-value
for a contingency table. Here's how you can do it:


import scipy.stats as stats

# Create the contingency table
observed = [[20, 15],
            [10, 25],
            [15, 20]]

# Perform the chi-square test
chi2, p, _, _ = stats.chi2_contingency(observed)

# Set the significance level
alpha = 0.05

# Compare p-value to the significance level
if p < alpha:
    print(f"Chi-square statistic: {chi2}")
    print(f"P-value: {p}")
    print("Reject the null hypothesis. There is a significant association 
          between the groups and outcomes.")
else:
    print(f"Chi-square statistic: {chi2}")
    print(f"P-value: {p}")
    print("Fail to reject the null hypothesis. There is no significant
          association between the groups and outcomes.")


In this code:

1. We create the observed contingency table with the given data.
2. We use `stats.chi2_contingency` to perform the chi-square test, which calculates the 
          chi-square statistic and p-value.
3. We set the significance level (`alpha`) to 0.05, as specified.
4. We compare the p-value to the significance level and interpret the results accordingly.

The code will print whether to "Reject the null hypothesis" or "Fail to reject the null hypothesis"
based on the p-value and significance level. In this case, the result will depend on the 
    specific values in your contingency table. If the p-value is less than 0.05, it suggests a
significant association between the groups and outcomes; otherwise, there is no significant association.

    
    
    
    
    
    
    
    







Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
population who smoke.




Ans:

To calculate the 95% confidence interval for the true proportion of individuals in the population
          who smoke, you can use the formula for the confidence interval for a proportion. The formula is:

          Confidence Interval = ^p ± Z⋅sqrt^p(1− ^p)/n



Where:
-^p is the sample proportion (the proportion of smokers in the sample).
- Z is the critical value from the standard normal distribution corresponding to
the desired confidence level (for 95% confidence, it's approximately 1.96).
- n is the sample size.

In your case, ^p  is the proportion of smokers in your sample, which is  60/500 , Z is 1.96
         for a 95% confidence level, and n is 500.

Let's calculate it in Python:


import math

# Sample size
n = 500

# Sample proportion
p_hat = 60 / n

# Z-score for 95% confidence level
z = 1.96

# Calculate the margin of error
margin_of_error = z * math.sqrt((^p * (1 - ^p)) / n)

# Calculate the confidence interval
lower_bound = ^p - margin_of_error
upper_bound = ^p + margin_of_error

# Print the confidence interval
print(f"95% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")


This Python code will calculate and print the 95% confidence interval for the true
proportion of individuals in the population who smoke.










Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation
of 12 using Python. Interpret the results.



Ans:


To calculate the 90% confidence interval for a sample of data with a mean
    of 75 and a standard deviation of 
12 using Python, you can use the `scipy.stats` library, which provides functions for 
various statistical calculations. Specifically, you can use the `norm.interval` 
function to calculate the confidence interval for a normal distribution. 
        Here's how you can do it:

import scipy.stats as stats

# Sample statistics
sample_mean = 75
sample_stddev = 12
confidence_level = 0.90

# Sample size (you need this to calculate the standard error)
sample_size =  # Replace with your actual sample size

# Calculate the standard error of the mean (SEM)
sem = sample_stddev / (sample_size**0.5)

# Calculate the margin of error (MOE) using the z-score for the desired confidence level
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_score * sem

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Interpretation
print(f"90% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")


Make sure to replace `sample_size` with the actual size of your sample.
        This code will calculate the 90% confidence interval
     for the given data and print the results.

Interpretation:
The 90% confidence interval for the population mean, based on the sample data with a mean of 75
 and a standard deviation of 12, is (xx.xx, xx.xx). This means that we are 90% confident 
   that the true population mean falls within this interval. In other words, if we were to take many 
      random samples and calculate their confidence intervals in the same way,
  we would expect about 90% of those intervals to contain the true population mean.










Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.




Ans:


To plot the chi-square distribution with 10 degrees of freedom in Python, you can use the `scipy.stats`
library to create the distribution and matplotlib for plotting. You can also shade the area corresponding
    to a chi-square statistic of 15 using matplotlib's `fill_between` function. Here's how you can do it:


import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the degrees of freedom
df = 10

# Create a range of x values for the chi-square distribution
x = np.linspace(0, 30, 1000)

# Calculate the probability density function (PDF) for the chi-square distribution
pdf = stats.chi2.pdf(x, df)

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x, pdf, label=f'Chi-Square ({df} degrees of freedom)', color='blue')

# Shade the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(0, 15, 1000)
pdf_fill = stats.chi2.pdf(x_fill, df)
plt.fill_between(x_fill, pdf_fill, color='lightcoral', alpha=0.5, label='Area for chi-square stat of 15')

# Label the axes and add a legend
plt.xlabel('Chi-Square Statistic')
plt.ylabel('Probability Density')
plt.title(f'Chi-Square Distribution with {df} Degrees of Freedom')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()


This code will create a plot of the chi-square distribution with 10 degrees of freedom and shade 
    the area corresponding to a chi-square statistic of 15. 
 Make sure you have the necessary libraries (`numpy`, `matplotlib`, and `scipy`) 
installed in your Python environment.




                              
                              
                              
                              
                              
                              
                              
                              


Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520
preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who
prefer Coke.


Ans:
                              
To calculate a 99% confidence interval for the true proportion of people 
in the population who prefer Coke,
you can use the formula for the confidence interval for a proportion. The formula is:

Confidence Interval  = ^p ± Z⋅sqrt^p(1− ^p)/n

Where:
- ^p is the sample proportion (520/1000 in this case).
- Z is the critical value for the desired confidence level (99% in this case). 
    You can find this value using a standard normal distribution table or a calculator. 
     For a 99% confidence level, Z ≈ 2.576 ).
- n is the sample size (1000 in this case).

Let's calculate the confidence interval step by step:

1. Calculate ^p:
   = ^p 520/1000 = 0.52 

2. Find the critical value \(Z\) for a 99% confidence level. You can use a standard normal
distribution table or a calculator. For a 99% confidence level, (Z ≈ 2.576 ).

3. Calculate the standard error (SE):
    SE = sqrt{0.52 .(1 - 0.52)}/{1000}

4. Calculate the margin of error (MOE):
    MOE = Z .SE 

5. Calculate the confidence interval:
   Confidence Interval = ^p ± MOE 

Now, plug in the values:

Confidence Interval = 0.52 ± 2.576 . sqrt{0.52 (1 - 0.52)}/{1000}

Calculating this, you will get the confidence interval:

Confidence Interval ≈ (0.484, 0.556) 

So, with 99% confidence, the true proportion of people in the population who prefer
Coke is estimated to be between 48.4% and 55.6%.
    
    
    
    
    
    
    
    
    
    
    
    







Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the
coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to
determine if the observed frequencies match the expected frequencies of a fair coin.
Use a significance level of 0.05.




Ans:
    

To conduct a chi-square goodness of fit test to determine if the observed frequencies
match the expected frequencies of a fair coin, you'll need to follow these steps:

**Step 1: Define your hypotheses:**

- Null Hypothesis (H0): The coin is fair, and the observed
frequencies match the expected frequencies.
- Alternative Hypothesis (Ha): The coin is biased towards tails, and the observed 
frequencies do not match the expected frequencies.

**Step 2: Set your significance level:**

You've already specified a significance level of 0.05.

**Step 3: Collect data:**

You have observed 45 tails out of 100 flips.

**Step 4: Determine expected frequencies:**

For a fair coin, the expected probability of getting tails in a single flip is 0.5. 
Therefore, the expected number of tails in 100 flips of a fair coin is:

Expected Tails = 0.5 * 100 = 50

**Step 5: Calculate the chi-square test statistic:**

The chi-square test statistic is calculated using the following formula:

χ² = Σ [(Observed frequency - Expected frequency)² / Expected frequency]

In your case, it would be:

χ² = [(45 - 50)² / 50] + [(55 - 50)² / 50]

Calculate this value:

χ² = [(45 - 50)² / 50] + [(55 - 50)² / 50] = (25/50) + (25/50) = 0.5 + 0.5 = 1

**Step 6: Determine the degrees of freedom (df):**

Degrees of freedom for a goodness of fit test is calculated as (number of categories - 1).
In this case, you have 2 categories (heads and tails), so df = 2 - 1 = 1.

**Step 7: Find the critical chi-square value:**

At a significance level of 0.05 and with 1 degree of freedom, you can find the critical 
chi-square value from a chi-square distribution table or calculator. 
The critical chi-square value is approximately 3.841.

**Step 8: Compare the calculated chi-square value to the critical chi-square value:**

Compare χ² = 1 to the critical chi-square value of 3.841.

Since χ² (1) < critical chi-square value (3.841), you fail to reject the null hypothesis.

**Step 9: Make a decision:**

At a significance level of 0.05, you do not have enough evidence to conclude that the coin is 
biased towards tails. Therefore, you cannot reject the null hypothesis. In other words, the
observed frequencies do not significantly differ from the expected frequencies for a fair coin.

**Step 10: Conclusion:**

Based on the chi-square goodness of fit test, there is no significant evidence to suggest 
that the coin is biased towards tails.
 
          
          
          
          




          
          
          
          
          
          




Q9. A study was conducted to determine if there is an association between smoking status (smoker or
non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below.
Conduct a chi-square test for independence to determine if there is a significant association between
smoking status and lung cancer diagnosis.


        Lung Cancer: Yes    Lung Cancer: No
Smoker     60                 140
Non-smoker 30                 170

Use a significance level of 0.05.




Ans:
To determine if there is a significant association between smoking status and lung cancer
diagnosis, you can conduct a chi-square test for independence using the given contingency table. 
The null hypothesis (H0) is that smoking status and lung cancer diagnosis are independent, 
while the alternative hypothesis (H1) is that they are not independent.

Here's how you can perform the chi-square test step by step:

**Step 1: Set up the hypotheses**

- Null Hypothesis (H0): Smoking status and lung cancer diagnosis are independent.
- Alternative Hypothesis (H1): Smoking status and lung cancer diagnosis are not independent.

**Step 2: Create an Observed Frequency Table**

        Lung Cancer: Yes    Lung Cancer: No    Row Total
Smoker     60                 140                  200
Non-smoker 30                 170                  200
Column Total 90                310                  400

**Step 3: Calculate Expected Frequencies**

To calculate the expected frequencies for each cell under the assumption 
of independence, you can use the formula:

Expected Frequency = (Row Total * Column Total) / Grand Total

Let's calculate the expected frequencies:

- Expected Frequency for "Smoker - Lung Cancer: Yes" cell:
  = (200 * 90) / 400 = 45

- Expected Frequency for "Smoker - Lung Cancer: No" cell:
  = (200 * 310) / 400 = 155

- Expected Frequency for "Non-smoker - Lung Cancer: Yes" cell:
  = (200 * 90) / 400 = 45

- Expected Frequency for "Non-smoker - Lung Cancer: No" cell:
  = (200 * 310) / 400 = 155

**Step 4: Calculate the Chi-Square Statistic**

Now, calculate the chi-square statistic using the formula:

Chi-Square = Σ [(Observed Frequency - Expected Frequency)^2 / Expected Frequency]

Sum this value over all cells in the table.

Chi-Square = [(60 - 45)^2 / 45] + [(140 - 155)^2 / 155] + [(30 - 45)^2 / 45] + [(170 - 155)^2 / 155]

Chi-Square ≈ 5.556

**Step 5: Determine the Degrees of Freedom**

Degrees of Freedom (df) = (Number of Rows - 1) * (Number of Columns - 1)
df = (2 - 1) * (2 - 1) = 1

**Step 6: Determine the Critical Value**

At a significance level of 0.05 and 1 degree of freedom, you can consult
          a chi-square distribution table
 or use a chi-square calculator to find the critical value.       
For df=1 and α=0.05, the critical value is approximately 3.841.

**Step 7: Compare the Chi-Square Statistic to the Critical Value**

Chi-Square Statistic (5.556) > Critical Value (3.841)

**Step 8: Make a Decision**

Since the Chi-Square Statistic is greater than the Critical Value,
you reject the null hypothesis (H0).

**Step 9: Interpret the Result**

There is a significant association between smoking status and lung cancer diagnosis 
at a significance level of 0.05. In other words, smoking status and lung cancer
diagnosis are not independent; they are related.
















Q10. A study was conducted to determine if the proportion of people who prefer milk chocolate, dark
chocolate, or white chocolate is different in the U.S. versus the U.K. A random sample of 500 people from
the U.S. and a random sample of 500 people from the U.K. were surveyed. The results are shown in the
contingency table below. Conduct a chi-square test for independence to determine if there is a significant
association between chocolate preference and country of origin.
Use a significance level of 0.01.

              Milk Chocolate        Dark Chocolate   White Chocolate
U.S. (n=500)   200                     150             150
U.K. (n=500)   225                     175             100




Ans:



To determine if there is a significant association between chocolate preference and 
the country of origin (U.S. or U.K.), you can conduct a chi-square test for independence.
Here are the steps to perform this test:

Step 1: Set up hypotheses
- Null Hypothesis (H0): There is no association between chocolate preference and country of origin.
- Alternative Hypothesis (Ha): There is an association between chocolate preference and country of origin.

Step 2: Calculate the expected frequencies for each cell in the contingency table.
Expected frequencies are what we would expect to observe if there were no association
between the two variables. The formula for calculating the expected frequency for a cell is:

Expected Frequency = (Row Total * Column Total) / Grand Total

Calculate the expected frequencies:

- Expected frequency for U.S. preference for Milk Chocolate: (500 * (200+225)) / 1000 = 212.5
- Expected frequency for U.S. preference for Dark Chocolate: (500 * (150+175)) / 1000 = 162.5
- Expected frequency for U.S. preference for White Chocolate: (500 * (150+100)) / 1000 = 137.5
- Expected frequency for U.K. preference for Milk Chocolate: (500 * (200+225)) / 1000 = 287.5
- Expected frequency for U.K. preference for Dark Chocolate: (500 * (150+175)) / 1000 = 237.5
- Expected frequency for U.K. preference for White Chocolate: (500 * (150+100)) / 1000 = 202.5

Step 3: Calculate the chi-square statistic
The chi-square statistic is calculated using the formula:

chi-square = Σ [(O - E)^2 / E]

Where:
- O = Observed frequency in a cell
- E = Expected frequency in a cell
- Σ = Summation across all cells

Calculate the chi-square statistic for each cell and then sum them up.

chi-square = [(200 - 212.5)^2 / 212.5] + [(150 - 162.5)^2 / 162.5] + [(150 - 137.5)^2 / 137.5]
+ [(225 - 287.5)^2 / 287.5] + [(175 - 237.5)^2 / 237.5] + [(100 - 202.5)^2 / 202.5]

chi-square ≈ 36.88

Step 4: Determine the degrees of freedom
Degrees of freedom (df) in a chi-square test for independence are calculated as:

df = (number of rows - 1) * (number of columns - 1)

In this case, df = (2 - 1) * (3 - 1) = 1 * 2 = 2.

Step 5: Determine the critical chi-square value
With a significance level of 0.01 and 2 degrees of freedom, you can look up the
critical chi-square value from a chi-square table or calculator.
For a significance level of 0.01 and 2 df, the critical chi-square value is approximately 9.21.

Step 6: Compare the calculated chi-square value to the critical chi-square value
Since the calculated chi-square value (36.88) is greater than the critical
chi-square value (9.21), you reject the null hypothesis.

Step 7: Make a conclusion
There is sufficient evidence to conclude that there is a significant association between 
chocolate preference and country of origin at a significance level of 0.01.

In other words, it appears that chocolate preference is not independent of the country of origin,
and there is a significant difference in chocolate preferences between the U.S. and the U.K.





    
    
    
    
    
    
    
    




Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.





Ans:


To conduct a hypothesis test to determine if the population mean is significantly different from 70,
you can use a one-sample t-test since you have a sample mean, sample standard deviation,
and want to compare it to a known population mean (70). Here are the steps for the hypothesis test:

**Step 1: Set up the null and alternative hypotheses.**

The null hypothesis (H0) is that the population mean is equal to 70:
H0: μ = 70

The alternative hypothesis (H1) is that the population mean is not equal to 70:
H1: μ ≠ 70

**Step 2: Determine the significance level.**

The significance level (alpha, α) is given as 0.05.

**Step 3: Calculate the test statistic.**

You can use the t-test formula:
       t =  ˉx−μ/s/sqrt{n}


Where:
- ˉx is the sample mean (72).
- μ is the hypothesized population mean (70).
- s is the sample standard deviation (10).
- n is the sample size (30).

Substitute the values into the formula:
t = 72 - 70 10 /10 \sqrt{30}
t = 2/10 /sqrt{30}

Now, calculate t:
t ≈ 1.0954

**Step 4: Determine the critical values.**

Since this is a two-tailed test (H1: μ ≠ 70), you need to find the critical values for a 
          significance level of 0.05/2 = 0.025 in each tail.
You can use a t-distribution table or a calculator. With 29 degrees of freedom (n - 1), 
the critical t-values for a two-tailed test with α = 0.025 are approximately ±2.045.

**Step 5: Compare the test statistic to the critical values.**

Your calculated t-value is approximately 1.0954, and the critical t-values are ±2.045.

Since \(|1.0954| < 2.045\), you fail to reject the null hypothesis.

**Step 6: Make a decision and interpret the results.**

Based on the results of the hypothesis test, at the 0.05 significance level, there is not
enough evidence to conclude that the population mean is significantly different from 70.

In other words, you do not have sufficient evidence to suggest that the sample mean of 72 is
significantly different from the hypothesized population mean of 70 at the 0.05 significance level.










```
