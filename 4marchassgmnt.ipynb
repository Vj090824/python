```python
Q1. Load the "titanic" dataset using the load_dataset function of seaborn.
Use Plotly express to plot scatter plot for age and fare columns in the titanic dataset.

Ans:

To load the "titanic" dataset using the `load_dataset` function from Seaborn and create 
a scatter plot for the "age" and "fare" columns using Plotly Express, you can follow 
these steps. First, you need to import the necessary libraries and then perform the 
data loading and plotting:

import seaborn as sns
import plotly.express as px

# Load the Titanic dataset from Seaborn
titanic_data = sns.load_dataset("titanic")

# Create a scatter plot using Plotly Express
scatter_plot = px.scatter(titanic_data, x="age", y="fare", title="Scatter Plot of Age vs Fare")

# Show the plot
scatter_plot.show()


Make sure you have both Seaborn and Plotly installed in your Python environment. 
You can install them using pip if you haven't already:


pip install seaborn plotly


This code will load the "titanic" dataset, create a scatter plot of the "age" and "fare" 
columns, and display the plot using Plotly Express.
    
    
    
    
    
    
    
    
    
    
    

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.


Ans:

Python code to create a box plot using the Plotly Express library with the tips dataset. 
You'll need to have Plotly installed in your Python environment to run this code. 
If you haven't already installed Plotly, you can do so using pip:


pip install plotly


Here's a Python script to create a box plot using Plotly Express with the tips dataset:


import plotly.express as px
import seaborn as sns

# Load the tips dataset from seaborn
tips = sns.load_dataset("tips")

# Create a box plot using Plotly Express
fig = px.box(tips, x="day", y="total_bill", color="time", 
             title="Box Plot of Total Bill by Day and Time")
fig.show()

In this code:

1. We import the necessary libraries, including Plotly Express and Seaborn.
2. We load the "tips" dataset from Seaborn. This dataset contains information about restaurant tips.
3. We use Plotly Express to create a box plot (`px.box`) with "day" on the x-axis,
"total_bill" on the y-axis, and color-coded by "time."
4. Finally, we display the plot using `fig.show()`.

Make sure you have both Plotly and Seaborn installed in your Python environment before running
this code. You can adjust the plot's title and other styling options as needed.


    
    
    
    
    
    
    
    




Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex"
and y="total_bill" column in the tips dataset. Also, use the "smoker" column with 
the pattern_shape parameter and the "day" column with the color parameter.


Ans:

    
To create a histogram using the Plotly library for the "total_bill" column in the tips
dataset, with the "sex" column as the x-axis, the "smoker" column as the pattern_shape
parameter, and the "day" column as the color parameter, you can follow these steps in Python:


import plotly.express as px
import seaborn as sns

# Load the tips dataset from seaborn
tips = sns.load_dataset("tips")

# Create a histogram using Plotly Express
fig = px.histogram(
    tips,
    x="sex",
    y="total_bill",
    color="day",
    pattern_shape="smoker",
    title="Histogram of Total Bill by Sex, Colored by Day and Shaped by Smoker",
    labels={"sex": "Sex", "total_bill": "Total Bill"},
)

# Show the plot
fig.show()


In this code:

1. We import the necessary libraries, including `plotly.express`
for creating the Plotly plot and `seaborn` to load the tips dataset.
2. We load the tips dataset using `sns.load_dataset("tips")`.
3. We create the histogram using `px.histogram()`, specifying "sex" as the x-axis,
"total_bill" as the y-axis, "day" as the color parameter, and "smoker" as the 
pattern_shape parameter.
4. We set the title and labels for the plot.
5. Finally, we use `fig.show()` to display the plot.

This code will generate a histogram with the specified parameters for the "sex," 
"total_bill," "smoker," and "day" columns in the tips dataset.
    
    
    
    
    
    
    
    
    
    
    
    

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot,

using the "species" column for the color parameter.

Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.


Ans:
    
    
To create a scatter matrix plot using the iris dataset in Plotly with the "species" column 
for the color parameter and using only the specified columns, you can follow these steps. 
First, make sure you have the necessary libraries installed, including Plotly. 
You can install Plotly using pip if you haven't already:


pip install plotly


Here's the code to create the scatter matrix plot:


import plotly.express as px
import seaborn as sns
import pandas as pd

# Load the iris dataset
df = sns.load_dataset("iris")

# Create a scatter matrix plot using Plotly
scatter_matrix = px.scatter_matrix(
    df,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Iris Dataset Scatter Matrix",
    labels={"species": "Species"},
)

# Update the diagonal subplots to use histogram type
scatter_matrix.update_traces(diagonal_visible=False)
scatter_matrix.update_layout(diagonal=dict(type='histogram'))

# Show the plot
scatter_matrix.show()


In this code:

1. We load the iris dataset using the Seaborn library.
2. We use Plotly Express to create a scatter matrix plot (`px.scatter_matrix`) 
with the specified dimensions ("sepal_length", "sepal_width", "petal_length", 
                               "petal_width") and use the "species" column for the color parameter.
3. We update the diagonal subplots to show histograms instead of scatter plots for each variable.
4. Finally, we display the scatter matrix plot using `scatter_matrix.show()`.    
    
    
    
    








Q5. What is Distplot? Using Plotly express, plot a distplot.



Ans:
    
In Plotly Express, the `distplot` function is used to create a distribution plot, also known as
a histogram with a kernel density estimate (KDE) 
overlaid on top of it. A distribution plot is a great way to visualize the distribution of a 
single variable, showing how the values are spread out and the approximate 
shape of the underlying data.

Here's how you can create a distribution plot using Plotly Express:


import plotly.express as px

# Sample data
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 7]

# Create a distribution plot
fig = px.histogram(data, x=data, nbins=10, opacity=0.7, title="Distribution Plot")

# Add a KDE overlay
fig.update_traces(histnorm='probability density', marker_line_width=0, marker_line_color="black")

# Show the plot
fig.show()

In this example:

1. We import `plotly.express` as `px`.
2. We define some sample data in the `data` list.
3. We create a distribution plot using `px.histogram()`, passing in the data, the variable 
to plot (`x=data`), and the number of bins (`nbins=10`). We also set the opacity of the 
bars and give the plot a title.
4. We use `fig.update_traces()` to add a KDE overlay to the histogram. Setting 
`histnorm='probability density'` scales the histogram bars to represent the probability density,
and `marker_line_width=0` and `marker_line_color="black"` remove the outline around the bars.
5. Finally, we display the plot using `fig.show()`.

This code will generate a distribution plot with a histogram and KDE overlaid on top of it,
allowing you to visualize the distribution of the `data` variable. You can customize the data 
and plot appearance according to your specific needs.   
    
    
    
    







```
